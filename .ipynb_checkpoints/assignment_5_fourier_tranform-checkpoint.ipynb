## 2. Spectral convolution

```python
def shift(x, k, l, boundary):
    if x.ndim == 2:
        color = 1
    else :
        color = 3
    n1 = np.shape(x)[0]
    n2 = np.shape(x)[1]
    xshifted = np.zeros((n1,n2,color))
    irange = np.mod(np.arange(n1) + k, n1)
    jrange = np.mod(np.arange(n2) + l, n2)
    # firstly move upward then move rightward
    xshifted = x[irange, :][:, jrange]
    if boundary == 'periodical':
        pass
    elif boundary is 'extension':
        m = n1 - k if k > 0 else -k-1
        n = n2 - l if l > 0 else -l-1
        if k != 0:
            xshifted[m::np.sign(k),:,:] = np.tile(xshifted[m-np.sign(k):m-np.sign(k)+1,:,:],(np.sign(k)*k,1,1))
        if l != 0:
            xshifted[:,n::np.sign(l),:] = np.tile(xshifted[:,n-np.sign(l):n-np.sign(l)+1,:],(1,np.sign(l)*l,1))
    elif boundary == 'zero-padding':
        period = xshifted
        xshifted = np.zeros_like(period)
        m = n1 - k if k > 0 else -k-1
        n = n2 - l if l > 0 else -l-1  
        sign_k = np.sign(k) if k != 0 else 1 
        sign_l = np.sign(l) if l != 0 else 1
        if k == 0:
            m = n1
        if l == 0:
            n = n2
        if color == 3:
            xshifted[:m:sign_k,:n:sign_l,:] = period[:m:sign_k,:n:sign_l,:]
        else:
            xshifted[:m:sign_k,:n:sign_l] = period[:m:sign_k,:n:sign_l]
    # mirror
    else:
        m = n1 - k if k > 0 else -k
        n = n2 - l if l > 0 else -l
        add_k = 1 if k < 0 else 0
        add_l = 1 if l < 0 else 0
        if color == 3:
            if k != 0:
                xshifted[m::np.sign(k),:,:] = xshifted[min(m,m-k):max(m,m-k) + add_k,:,:][::-np.sign(k),:,:]
            if l != 0:
                xshifted[:,n::np.sign(l),:] = xshifted[:,min(n,n-l):max(n,n-l) + add_l ,:][:,::-np.sign(l),:]
        else:
            if k != 0:
                xshifted[m::np.sign(k),:] = xshifted[min(m,m-k):max(m,m-k) + add_k,:][::-np.sign(k),:]
            if l != 0:
                xshifted[:,n::np.sign(l)] = xshifted[:,min(n,n-l):max(n,n-l) + add_l][:,::-np.sign(l)]
            
    return xshifted
```

```python
def convolve(x, nu, boundary= 'periodical'):
    xconv = np.zeros(x.shape)
    s1 = int((nu.shape[0] - 1) / 2)
    s2 = int((nu.shape[1] - 1) / 2)
    for k in range(-s1, s1+1):
        for l in range(-s2, s2+1):
            xconv += nu[k+s1,l+s2]*shift(x,-k,-l,boundary)
    return xconv
```

```python
def kernel(name, tau=1, eps=1e-3):
    if name == 'gaussian':
        s1 = 0
        while True:
            if np.exp(-(s1**2)/(2*tau)) < eps:
                break
            s1 += 1
        s1 = s1-1
        s2 = s1
        i = np.arange(-s1,s1+1) #-3 ~ 3
        j = np.arange(-s2,s2+1) #-3 ~ 3 
        ii, jj = np.meshgrid(i, j, sparse=True,indexing='ij')
        nu = np.exp(-(ii**2 + jj**2) / (2*tau**2))
        nu[nu < eps] = 0
        nu /= nu.sum()
    elif name == 'exponential':
        if tau == 0:
            s1 = 1
            s2 =1
        else:
            s1 = 0
            while True:
                if np.exp(-(s1)/(tau)) < eps:
                    break
                s1 += 1
            s1 = s1-1
            s2 = s1
            
        i = np.arange(-s1,s1+1) #-20 ~ 20
        j = np.arange(-s2,s2+1) #-20 ~ 20
        ii, jj = np.meshgrid(i, j, sparse=True,indexing='ij')
        if tau == 0: 
            tau = 1e-3
        nu = np.exp(-(np.sqrt(ii**2+jj**2))/tau)
        nu[nu < eps] = 0
        nu /= nu.sum()
    elif name.startswith('box'):
        if name.endswith('1'):
            nu = np.zeros((2*tau+1, 1))
            nu[:,0] = 1/(2*tau+1)
        elif name.endswith('2'):
            nu = np.zeros((1, 2*tau+1))
            nu[0,:] = 1/(2*tau+1)
        else:
            s1 = tau
            s2 = tau
            i = np.arange(-s1,s1+1) #-1 ~ 1
            j = np.arange(-s2,s2+1) #-1 ~ 1
            ii, jj = np.meshgrid(i, j, sparse=True,indexing='ij')
            nu = np.exp(0*(ii+jj))
            nu[nu < eps] = 0
            nu /= nu.sum()
    elif name == 'motion':
        nu = np.load('./assets/motionblur.npy')
    elif name.endswith('forward'):
        if name.startswith('1',4,5):
            nu = np.zeros((3, 1))
            nu[1, 0] = 1
            nu[2, 0] = -1
        elif name.startswith('2',4,5):
            nu = np.zeros((1, 3))
            nu[0, 1] = 1
            nu[0, 2] = -1
        else:
            raise ValueError('invalid kernel')
    elif name.endswith('backward'):
        if name.startswith('1',4,5):
            nu = np.zeros((3, 1))
            nu[0, 0] = 1
            nu[1, 0] = -1
        elif name.startswith('2',4,5):
            nu = np.zeros((1, 3))
            nu[0, 0] = 1
            nu[0, 1] = -1
        else: 
            raise ValueError('invalid kernel')
    elif name.startswith('laplacian'):
        if name.endswith('1'):
            nu = np.zeros((3, 1))
            nu[0, 0] = 1
            nu[1, 0] = -2
            nu[2, 0] = 1 
        elif name.endswith('2'):
            nu = np.zeros((1, 3))
            nu[0, 0] = 1
            nu[0, 1] = -2
            nu[0, 2] = 1
        else: 
            raise ValueError('invalid kernel')
    else:
        raise ValueError('invalid kernel')
    return nu
```

```python
def kernel2fft(nu, n1, n2, separable = None):

    s1 = int((nu.shape[0] - 1) / 2)
    s2 = int((nu.shape[1] - 1) / 2)
    
    tmp = np.zeros((n1,n2))
    tmp[:s1+1, :s2+1] = nu[s1:2*s1+1, s2:2*s2+1]
    tmp[n1-s1:n1, :s2+1] = nu[:s1,s2:2*s2+1]
    tmp[n1-s1:n1,n2-s2:n2] = nu[:s1,:s2]
    tmp[:s1+1,n2-s2:n2] = nu[s1:2*s1+1,:s2]
    
    lbd = npf.fft2(tmp)
    
    return lbd
```

$\textbf{9.}$

```python
def convolvefft(x, lbd):
    return np.real(npf.ifftn(npf.fftn(x)*lbd[:,:,None]))
```

```python
img = plt.imread('assets/house.png')
overall_time_normal = np.zeros((5))
overall_time_fft = np.zeros((5))
overall_time_normal_sep = np.zeros((5))
overall_time_fft_sep = np.zeros((5))
n = 100 #100
for tau in range(0,5):
    for it in range(n):
        # Non-separable
        normal_ker= kernel('box',tau)
        fft_kernel = kernel2fft(normal_ker,img.shape[0],img.shape[1])
        start_time = time.time()
        res_1 = convolve(img,normal_ker,boundary = 'periodical')
        elapse_time = (time.time() - start_time)*1000
        overall_time_normal[tau] += elapse_time
        print("tau = " + str(tau)+" :non-separable general convolve time = " + str(elapse_time) + " ms")
        start_time = time.time()
        res_2 = convolvefft(img,fft_kernel)
        elapse_time = (time.time() - start_time)*1000
        overall_time_fft[tau] += elapse_time
        print("tau = " + str(tau)+" :non-separable FFT convolve time = " + str(elapse_time) + " ms")
        
        # Separable
        sep_ker1 = kernel('box1',tau)
        sep_ker2 = kernel('box2',tau)
        start_time = time.time()
        res_31 = convolve(img,sep_ker1,boundary = 'periodical')
        res_32 = convolve(res_31,sep_ker2,boundary = 'periodical')
        elapse_time = (time.time() - start_time)*1000
        overall_time_normal_sep[tau] += elapse_time
        print("tau = " + str(tau)+" :Separable general convolve time = " + str(elapse_time) + " ms")
        
        fft_sep_ker1 = kernel2fft(sep_ker1,img.shape[0],img.shape[1])
        fft_sep_ker2 = kernel2fft(sep_ker2,img.shape[0],img.shape[1])
        fft_sep = np.multiply(fft_sep_ker1,fft_sep_ker2)
        start_time = time.time()
        res_42 = convolvefft(img,fft_sep)
        #res_41 = convolvefft(img,fft_sep_ker1)
        #res_42 = convolvefft(res_41,fft_sep_ker2)
        elapse_time = (time.time() - start_time)*1000
        overall_time_fft_sep[tau] += elapse_time
        print("tau = " + str(tau)+" :Separable FFT convolve time = " + str(elapse_time) + " ms")
        
overall_time_normal /= n
overall_time_fft /= n
overall_time_normal_sep /= n
overall_time_fft_sep /= n
plt.figure()
plt.title('Separable filter')
plt.plot(overall_time_normal_sep,'b', label="General")
plt.plot(overall_time_fft_sep,'C1',label="FFT")
plt.xlabel('tau', fontsize=18)
plt.ylabel('Time', fontsize=16)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
plt.figure()
plt.title('Non-separable filter')
plt.plot(overall_time_normal,'b',label="General")
plt.plot(overall_time_fft,'C1',label="FFT")
plt.xlabel('tau', fontsize=18)
plt.ylabel('Time', fontsize=16)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

* Spectral convolution is less relate to the value of tau which means that when tau increase, the time of spectral convolution doesn't increase. On the other hand, spatial convolution does based on the value of tau.
* Using separable kernel is more favorable when using spatial convolution since it can save lots of time. On the other hand, it doesn't save lots of time when using spectral kernel.

* According to the above plots, spatial covolution is better than the spectral convolutiom when tau is very small.

$\textbf{14.}$

$
\begin{cases}
H(x) = \nu * x = F^{-1}(F \nu \times F x) \\ 
H^*(x) = \mu * x = F^{-1}(F \mu \times F x)
\end{cases}
$

$
(F^{-1}(F \nu \times F x))^* = F^{-1}(F \mu \times F x)
$

$F^{-1}=F^H$. And, $F^T=F$


$1)$ The left term

$
(F\nu F x)^*(F^H)^* \\
= x^* F^* \nu^* F^* F^T \\
= x^* F^* \nu^* F^{-1} F^T \\
= x^* F^* \nu^* \\
= (\nu F x)^*
$

$2)$ The right term

$
F^{-1} F \mu F x \\
= \mu F x
$



$
(F \nu x)^* = F \mu x
$


$
\mu(x,y)=\nu(-x,-y)
$

```python
def nu2mu(nu):
    return np.fliplr(np.flipud(nu))
```

```python
x = plt.imread('assets/house.png')
y = plt.imread('assets/map.png')
nu = kernel('gaussian',5,1e-3)
mu = nu2mu(nu)
np.isclose(np.multiply(convolve(x,nu),y).sum(),np.multiply(x,convolve(y,mu)).sum())
```

$\textbf{15.}$

$\mathbf{\because \nu(x,y) = F^{-1}(F\mu)^*}$

$\mathbf{\therefore F\nu(x,y) = (F\mu)^*}$

$\textbf{The frequency response of}$ $\nu$ $\textbf{is the conjugate of}$ $\mu$ $\textbf{frequency response}$

```python
lbd_nu = kernel2fft(kernel('gaussian',5,1e-3),x.shape[0],x.shape[1],None)
lbd_mu = np.conj(lbd_nu)
convolvefft(x,lbd_nu)
np.isclose(np.multiply(convolvefft(x,lbd_nu),y).sum(),np.multiply(x,convolvefft(y,lbd_mu)).sum())
```
